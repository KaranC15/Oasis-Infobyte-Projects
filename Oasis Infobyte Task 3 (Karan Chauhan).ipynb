### Checking columns with 'object' data type and encoding them.

```python
xo = df.select_dtypes(include='object')
xo
```

```python
x = df.drop('price',axis=1)
y = df.price
```

```python
lb = LE()
for col in xo.columns:
    x[col] = lb.fit_transform(x[col].values)
x
```

### Splitting data into training and testing sets

```python
x_train,x_test,y_train,y_test = TTS(x,y)
```

### Introducing linear regression model

```python
lr = LR()
lr.fit(x_train,y_train)
```

### Checking for traning and testing scores

```python
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
y_pred = lr.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
def SSE(y_test,y_pred):
    return sum((y_test-y_pred)**2)
def MAE(y_test,y_pred):
    return np.mean(np.abs(y_test-y_pred))
def MSE(y_test,y_pred):
    return np.mean((y_test-y_pred)**2)
def RMSE(y_test,y_pred):
    return (np.mean((y_test-y_pred)**2))**0.5
```

```python
print('Sum of Squared Error -',SSE(y_test,y_pred))
print('Mean Absolute Error -',MAE(y_test,y_pred))
print('Mean Squared Error -',MSE(y_test,y_pred))      
print('Root Mean Squared Error -',RMSE(y_test,y_pred))      
```
